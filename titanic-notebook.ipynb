## Import input csv files using pandas

```python
train=pd.read_csv("../input/titanic/train.csv")
test=pd.read_csv("../input/titanic/test.csv")
#full_data=pd.concat([train, test], axis=1, sort=False)
full_data=train.append(test, sort=True)
full_data.tail(10)
```

### set PassengerId as a Index

```python
full_data.set_index(["PassengerId"], inplace=True)
full_data
```

### Convert the numerical values into string

```python
passengerClass=["Upper", "Middle", "Lower"]
for i, passClass in zip(range(1, 4), passengerClass):
    full_data["Pclass"]=full_data["Pclass"].replace(i, passClass)

full_data.head()
```

```python
for i, choice in zip(range(1,full_data.shape[0] + 1), full_data["Cabin"].isna()):
    if not choice:
        full_data["Cabin"][i]=full_data["Cabin"][i].replace(full_data["Cabin"][i], full_data["Cabin"][i][0])
    else:
        pass
```

```python
full_data.head()
```

```python
full_data.drop(["Name", "Ticket"], axis=1, inplace=True)
full_data.head()
```

```python
full_data["FamilySize"]=full_data["Parch"]+full_data["SibSp"]+1
full_data.head()
```

```python
full_data.drop(["Parch", "SibSp"],axis=1, inplace=True)
full_data.head()
```

```python
dum_full_data=pd.get_dummies(full_data)
dum_full_data.head()
```

```python
dum_full_data["Age"]=dum_full_data["Age"].fillna(dum_full_data["Age"].mean())
dum_full_data["Fare"]=dum_full_data["Fare"].fillna(dum_full_data["Fare"].mean())
##dum_full_data["Survived"]=dum_full_data["Survived"].fillna(dum_full_data["Survived"].median())
dum_full_data.tail()
```

```python
train_data=dum_full_data.loc[1:891,:]
test_data=dum_full_data.loc[892:,:]
train_data.tail(20)
```

```python
test_data.head()
```

```python
X_train=train_data.drop(["Survived", "Fare"], axis=1)
y_train=train_data["Survived"]
X_test=test_data.drop(["Survived", "Fare"], axis=1)
```

```python
X_train.head()
```

```python
gender_submission=pd.read_csv("../input/titanic/gender_submission.csv")
mysubmission=gender_submission.copy()
gender_submission.head()
```

```python
y_test=gender_submission["Survived"]
```

```python
logreg = LogisticRegression(solver="saga", max_iter=100, multi_class="auto")
logreg.fit(X_train, y_train)
y_pred1 = logreg.predict(X_test)
#acc_log = round(logreg.score(X_train, y_train) * 100, 2)
#acc_log
r2_score(y_test, y_pred1)
```

```python
mysubmission["Survived"]=y_pred1.astype(int)
mysubmission.to_csv("mysubmission1.csv", index=False)
```

```python
mlp = MLPRegressor(hidden_layer_sizes=(10, 10, 10, 10),random_state=52,
                   learning_rate="adaptive", activation="relu", solver="lbfgs",
                  early_stopping=True)
mlp.fit( X_train , y_train )
y_pred2 = mlp.predict(X_test)
#acc_mlp = round(mlp.score(X_train, y_train) * 100, 2)
#acc_mlp
r2_score(y_test, y_pred2)
```

```python
depth_range = range(1, 50, 2)
minsplit_range = range(5, 100, 5)
minleaf_range = range(2, 100, 5)
```

```python
parameters=dict(max_depth=depth_range,
               min_samples_split=minsplit_range,
               min_samples_leaf=minleaf_range)
parameters
```

```python
kfold=StratifiedKFold(n_splits=5, random_state=42, shuffle=True)
kfold
```

```python
clf=DecisionTreeClassifier(random_state=2018)
cv=GridSearchCV(clf, param_grid=parameters, cv=kfold, scoring="roc_auc")
cv.fit(X_train, y_train)
print(cv.best_params_)
print(cv.best_score_)
```

```python
y_pred3=cv.predict(X_test)
```

```python
mysubmission2=gender_submission.copy()
mysubmission2["Survived"]=y_pred3.astype(int)
y_pred3
```

```python
mysubmission2.to_csv("mysubmission2.csv", index=False)
```
